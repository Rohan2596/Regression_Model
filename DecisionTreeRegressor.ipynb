```python
X=df[['ENGINESIZE', 'CYLINDERS', 'FUELTYPE','FUELCONSUMPTION_COMB','FUELCONSUMPTION_COMB_MPG']].values
X[0:5]
```

```python
y=df[['CO2EMISSIONS']].values
y[0:5]
```

```python
from sklearn import preprocessing
le_fueltype=preprocessing.LabelEncoder()
le_fueltype.fit(['X','Z','E','D'])
X[:,2]=le_fueltype.transform(X[:,2])
X[0:5]
```

```python
from sklearn.tree import DecisionTreeRegressor
Decision_Tree=DecisionTreeRegressor()
Decision_model=Decision_Tree.fit(X,y)
Decision_predictions=Decision_model.predict(X)
Decision_predictions[0:25]
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,Decision_predictions)
```

```python
from sklearn.metrics import r2_score
r2_score(y,Decision_predictions)
```
